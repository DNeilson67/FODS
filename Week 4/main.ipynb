#### Data Cleaning My CSV 
##### Turning date to Year, Month and Day

```python
import re
def extract_date_parts(date_str):
    match = re.match(r'(\d{4})/(\d{2})/(\d{2})', date_str)
    if match:
        return match.groups()
    else:
        return None, None, None

df2[['Year', 'Month', 'Day']] = df2['date'].apply(lambda x: pd.Series(extract_date_parts(x)))
df2[['Year', 'Month', 'Day']] = df2[['Year', 'Month', 'Day']].apply(pd.to_numeric)
df2 = df2.drop(columns=['date'])
df2.head()
```

#### Data Cleaning movies_metadata.csv

```python
emdf = pd.read_csv("asset/movies_metadata.csv")
emdf
```

```python
import ast

emdf.dropna(subset = ["belongs_to_collection"], inplace = True)

emdf
collectionList = emdf['belongs_to_collection'].tolist()

idL = []
nameL = []
posterPthL = []
bckDrpPthL = []

collectionList

def verify(tempdict, col, array):
    try:
        array.append(tempdict[col])
    except:
        array.append("None")

for col in collectionList:
    try:
        tempdict = ast.literal_eval(col) ##Change string to dict
        verify(tempdict, "id", idL)
        verify(tempdict, "name", nameL)
        verify(tempdict, "poster_path", posterPthL)
        verify(tempdict, "backdrop_path", bckDrpPthL)
    except:
        idL.append("None")
        nameL.append("None")
        posterPthL.append("None")
        bckDrpPthL.append("None")

emdfDict = {
    "ID": idL,
    "name": nameL,
    "poster path":posterPthL,
    "backdrop path":bckDrpPthL
}

newDf = pd.DataFrame.from_dict(emdfDict)
newDf
```

```python
genreList = emdf["genres"].tolist()
genreIDL = []
genreNameL= []

extracted_data = []

for i in genreList:
    data_list = ast.literal_eval(i)
    extracted_data.append(data_list)

doubleArrayGenreID = []
doubleArrayNameGenre = []

for i in extracted_data:
    tempGroupID = []
    tempGroupName = []
    for j in i:
        tempGroupID.append(j["id"])
        tempGroupName.append(j["name"])
    doubleArrayGenreID.append(tempGroupID)
    doubleArrayNameGenre.append(tempGroupName)

emdfDict = {
    "ID": idL,
    "name": nameL,
    "poster path":posterPthL,
    "backdrop path":bckDrpPthL,
    "genre ID": doubleArrayGenreID,
    "genre name": doubleArrayNameGenre
}
    
newDf = pd.DataFrame.from_dict(emdfDict)
newDf
```
